```python
import pandas as pd
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)
df.info()
```

```python
mv = df.isnull().sum()
print(mv)
```

```python
# Find median age of males and females
male_age = df[df['Sex'] == 'male']['Age'].median()
female_age = df[df['Sex'] == 'female']['Age'].median()

# Fill missing ages for males and females separately
df.loc[(df['Age'].isnull()) & (df['Sex'] == 'male'), 'Age'] = male_age
df.loc[(df['Age'].isnull()) & (df['Sex'] == 'female'), 'Age'] = female_age
```

```python
# Get the title like Mr, Mrs, Miss, Master from the Name column
df['Title'] = df['Name'].apply(lambda x: x.split(',')[1].split('.')[0].strip())
```

```python
df['Title'] = df['Title'].replace(['Mlle', 'Ms'], 'Miss')
df['Title'] = df['Title'].replace(['Mme'], 'Mrs')

rare_titles = ['Dr', 'Rev', 'Col', 'Major', 'Countess', 'Lady', 'Jonkheer', 'Don', 'Dona', 'Capt', 'Sir']
for title in rare_titles:
    df['Title'] = df['Title'].replace(title, 'Rare')
```

```python
# Find median age for each Title
title_ages = df.groupby('Title')['Age'].median()

# Fill missing Age using median age for each Title
for title, age in title_ages.items():
    df.loc[(df['Age'].isnull()) & (df['Title'] == title), 'Age'] = age
```

```python
print(df.groupby('Title')['Survived'].mean())
```
